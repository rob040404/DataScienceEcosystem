### Exercise 1
Extract the required GDP data from the given URL using Web Scraping.


````python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
````

You can use Pandas library to extract the required table directly as a DataFrame. Note that the required table is the third one on the website, as shown in the image below.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/images/pandas_wbs_3.png">


````python
# Extract tables from webpage using Pandas. Retain table number 3 as the required dataframe.
tables = pd.read_html(URL)
df = tables[3]
#df.keys()
#df.head()

# Replace the column headers with column numbers
df.columns = range(df.shape[1])
df.columns

# Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)
df2 = df[[0, 2]]
df2
# Retain the Rows with index 1 to 10, indicating the top 10 economies of the world.
df3 = df2.iloc[1:11]
df3
# Assign column names as "Country" and "GDP (Million USD)"
df3.rename(columns = {0 : "Country", 2 : "GDP"}, inplace=True)
df3
````

La línea df.columns = range(df.shape[1]) está renombrando las columnas de un DataFrame de pandas de manera que sus nombres se conviertan en una secuencia de números consecutivos.

Explicación paso a paso:
df.columns:
Este atributo contiene los nombres de las columnas del DataFrame. Puedes acceder a él para ver los nombres actuales o para modificarlos.

df.shape[1]:
La propiedad shape del DataFrame devuelve una tupla que representa las dimensiones del DataFrame:

df.shape[0]: Número de filas.
df.shape[1]: Número de columnas.
Entonces, df.shape[1] devuelve el número total de columnas en el DataFrame.

range(df.shape[1]):
El comando range(n) genera una secuencia de números desde 0 hasta n-1. En este caso, range(df.shape[1]) generará una secuencia de números que va desde 0 hasta el número total de columnas menos 1. Por ejemplo, si el DataFrame tiene 5 columnas, esto será equivalente a range(5) o [0, 1, 2, 3, 4].

Asignación a df.columns:
Al asignar range(df.shape[1]) a df.columns, estás reemplazando los nombres actuales de las columnas con esta secuencia de números. Por ejemplo, si el DataFrame tiene 3 columnas con nombres como 'A', 'B', 'C', esta línea cambiará esos nombres por 0, 1, 2.

### Exercise 2
Modify the GDP column of the DataFrame, converting the value available in Million USD to Billion USD. Use the `round()` method of Numpy library to round the value to 2 decimal places. Modify the header of the DataFrame to `GDP (Billion USD)`.


````python
# Change the data type of the 'GDP (Million USD)' column to integer. Use astype() method.
df3 = df3.astype({"GDP" : "int32"})

# Convert the GDP value in Million USD to Billion USD
df3['GDP'] = df3['GDP']/1000
df3

# Use numpy.round() method to round the value to 2 decimal places.
df3['GDP'] = np.round(df3['GDP'], 2)
df3
# Rename the column header from 'GDP (Million USD)' to 'GDP (Billion USD)'
df3.rename(columns = {'GDP' : 'GDP (Billion USD)'}, inplace = True)
df3
````
